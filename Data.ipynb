```python
import pandas as pd
import yfinance as yf
```

```python
companies = pd.read_csv('ind_nifty500list.csv')
```

```python
companies.drop(columns=['Series', 'ISIN Code'], axis=0, inplace=True)
companies
```

```python
companies['volatility'] = 0
```

```python
companies
```

```python
for i in range(len(companies)):
    data = yf.download("".join([companies.Symbol[i], ".NS"]), start="2008-01-01", end="2009-12-31")
    if len(data) > 1:
        volatility = data['Adj Close'].std()/data['Adj Close'].mean()
        companies.at[i, 'volatility'] = volatility
```

```python
companies.sort_values(by=['volatility'], ascending=False, inplace=True)
```

```python
op_df = companies[:10]
```

```python
op_df.to_csv('companies.csv', index=False)
```

```python
op_df
```

```python
writer = pd.ExcelWriter('stocks_data.xlsx')

for i in range(len(op_df)):
    df = yf.download("".join([op_df.Symbol.iloc[i], ".NS"]), start="2010-01-01", end="2022-07-01")
    df.to_excel(writer, sheet_name = op_df.Symbol.iloc[i])    

writer.save()
```
